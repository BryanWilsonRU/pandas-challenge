```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df
```

## Player Count

* Display the total number of players


```python
#Find unique players (use SN [Screen Name])

totalPlayers=purchase_data_df.SN.nunique(dropna=False)

PC_summary_df = pd.DataFrame({"Total Number of Players": [totalPlayers]})
PC_summary_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Count unique values in column 'Item ID' of the dataframe
uniqueValues = purchase_data_df['Item ID'].nunique()

#Find average price (mean of Price column)
avg_price = purchase_data_df["Price"].mean()

#Find number of purchases (count of rows)
totalPurchases = purchase_data_df["Purchase ID"].count()

#Find total revenue (sum for Price column)
totalRevenue = purchase_data_df["Price"].sum()

#Summary table
PA_summary_df = pd.DataFrame({"Number of Unique Items": [uniqueValues],
                              "Average Price": avg_price,
                              "Total Purchases": totalPurchases,
                              "Total Revenue": totalRevenue})
PA_summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Get counts for Male, Female, Other
#Keep in mind that players may have made multiple purchases- make sure only to count unique individuals
#Percentage~ Male/Female/Other divided by total then multiply by 100

print("Gender Demographics")
males_df = purchase_data_df.loc[purchase_data_df["Gender"] == "Male", :]
uniqueMales = males_df['SN'].nunique()

pctMales=(uniqueMales/totalPlayers)*100
print(uniqueMales)
print(pctMales)

females_df = purchase_data_df.loc[purchase_data_df["Gender"] == "Female", :]
uniqueFems = females_df['SN'].nunique()

pctFems=(uniqueFems/totalPlayers)*100
print(uniqueFems)
print(pctFems)

other_nondisclosed_df = purchase_data_df.loc[purchase_data_df["Gender"] == "Other / Non-Disclosed", :]
uniqueOthers = other_nondisclosed_df['SN'].nunique()

pctOthers=(uniqueOthers/totalPlayers)*100
print(uniqueOthers)
print(pctOthers)

#Create a dataframe from a dictionary for use in next section of assignment
index = [0]
columns = ['Female', 'Male', 'Other / Non-Disclosed']
denominator_dict = {'Female': uniqueFems, 'Male': uniqueMales, 'Other / Non-Disclosed': uniqueOthers}      
purchasers_df = pd.DataFrame(denominator_dict, index = index, columns = columns) 
purchasers_df = purchasers_df.sum()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create a gropuby dataframe object for Gender
grouped_gender = purchase_data_df.groupby("Gender")

# Apply math to your group by object go get each value
# Purchase count by gender
gender_purchase_count = grouped_gender["Purchase ID"].count()

# Average purchase price by gender
gender_avg = grouped_gender["Price"].mean()

# Average purchase total by gender
gender_purchase_totals = grouped_gender["Price"].sum()

# Average purchase total per person by gender
gender_purchase_person = gender_purchase_totals/purchasers_df

#Summary table
gender_summary_df = pd.DataFrame({"Purchase Count (by Gender)": gender_purchase_count,
                              "Average Purchase Price (by Gender)": gender_avg,
                            "Average Purchase Total (by Gender)": gender_purchase_totals,
                            "Average Purchase per Person (by Gender)":gender_purchase_person})
gender_summary_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Find max age for final bin upper limit
purchase_data_df.describe()
```

```python
# Create and put data into bins

bins = [0, 9, 14, 19, 24, 29, 34, 39, 46]
age_labels = [" <10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data_df["Age Group"]= pd.cut(purchase_data_df["Age"], bins, labels=age_labels)
purchase_data_df.head()
```

```python
# Create a GroupBy object based upon "View Group"
age_group = purchase_data_df.groupby("Age Group")

# Get desired values within the GroupBy object

age_purchases = age_group["Purchase ID"].count()
age_avg_price = age_group["Price"].mean()
age_total_purchase = age_group["Price"].sum()

#Summary table
age_summary_df = pd.DataFrame({"Purchase Count": age_purchases,
                              "Average Purchase Price": age_avg_price,
                              "Total Purchase Value": age_total_purchase})
age_summary_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create a gropuby dataframe object for SN
grouped_SN = purchase_data_df.groupby("SN")

# Purchase count by SN
SN_purchase_count = grouped_SN["Purchase ID"].count()

# Average purchase price by gender
SN_avg = grouped_SN["Price"].mean()

# Average purchase total by gender
SN_purchase_totals = grouped_SN["Price"].sum()

#Summary table
SN_summary_df = pd.DataFrame({"Purchase Count": SN_purchase_count,
                              "Average Purchase Price": SN_avg,
                              "Total Purchase Value": SN_purchase_totals})
SN_summary_df

#Sort by total purchase in descending order
SN_summary_descending_df = SN_summary_df.sort_values("Total Purchase Value", ascending=False)
SN_summary_descending_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Reference multiple columns within a DataFrame
purchase_data_df[["Item ID", "Item Name", "Price"]].head()

# Create a gropuby dataframe object for Item ID
grouped_ID_and_Name = purchase_data_df.groupby(["Item ID","Item Name"])

# Purchase count by ID and Name
IDname_purchase_count = grouped_ID_and_Name["Purchase ID"].count()

# Item Price
IDname_price = grouped_ID_and_Name["Price"].max()

# Total Purchase Value
IDname_purchase_totals = grouped_ID_and_Name["Price"].sum()

#Summary table
IDname_summary_df = pd.DataFrame({"Purchase Count": IDname_purchase_count,
                              "Item Price": IDname_price,
                              "Total Purchase Value": IDname_purchase_totals})
IDname_summary_df


#Sort by purchase count in descending order
IDname_summary_descending_df = IDname_summary_df.sort_values("Purchase Count", ascending=False)
IDname_summary_descending_df.head()
```
